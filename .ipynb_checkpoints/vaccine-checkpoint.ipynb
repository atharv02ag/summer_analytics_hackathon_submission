```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_csv('./dataset and all/training_set_features.csv')
features.head(5)
```

```python
features.shape
```

```python
features.isnull().sum()
```

```python
feat = features.drop(columns=['employment_occupation','employment_industry','hhs_geo_region','census_msa','health_insurance','race','marital_status','household_adults','household_children'])
feat.head()
```

```python
feat = pd.get_dummies(feat,columns=['sex','income_poverty','employment_status','education','rent_or_own','age_group'],dtype = int)
feat.head()
```

```python
feat.isnull().sum()
```

```python
targets = pd.read_csv('./dataset and all/training_set_labels.csv')
targets.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = feat.drop(columns = ['respondent_id','xyz_concern','doctor_recc_xyz','opinion_xyz_vacc_effective','opinion_xyz_risk', 'opinion_xyz_sick_from_vacc'])
y = targets.drop(columns = ['respondent_id','xyz_vaccine'])

merged = pd.concat([X,y],ignore_index = True, axis = 1)
merged.columns = list(X.columns) + list(y.columns)

merged = merged.dropna()

X_train,X_test,y_train,y_test = train_test_split(merged.drop(columns = ['seasonal_vaccine']),merged['seasonal_vaccine'],test_size = 0.3,random_state = 42)
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import roc_auc_score

pred = model.predict_proba(X_test)
print(roc_auc_score(y_test,pred[:,1], average = 'macro'))
```

```python
X = feat.drop(columns = ['respondent_id','doctor_recc_seasonal','opinion_seas_vacc_effective','opinion_seas_risk','opinion_seas_sick_from_vacc'])
y = targets.drop(columns = ['respondent_id','seasonal_vaccine'])

merged = pd.concat([X,y],ignore_index = True, axis = 1)
merged.columns = list(X.columns) + list(y.columns)

merged = merged.dropna()

X_train,X_test,y_train,y_test = train_test_split(merged.drop(columns = ['xyz_vaccine']),merged['xyz_vaccine'],test_size = 0.3,random_state = 42)
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
pred = model.predict_proba(X_test)
print(roc_auc_score(y_test,pred[:,1], average = 'macro'))
```

```python
test_features = pd.read_csv('./dataset and all/test_set_features.csv')
print(test_features.shape)
```

```python
test_feat = test_features.drop(columns=['employment_occupation','employment_industry','hhs_geo_region','census_msa','health_insurance','race','marital_status','household_adults','household_children'])
test_feat = pd.get_dummies(test_feat,columns=['sex','income_poverty','employment_status','education','rent_or_own','age_group'],dtype = int)

test_feat.head(5)
```

```python
#for xyz
X_test = test_feat.drop(columns = ['respondent_id'])
X_train = feat.drop(columns = ['respondent_id'])
y_train = targets.drop(columns = ['respondent_id','seasonal_vaccine'])

merged = pd.concat([X_train,y_train],ignore_index = True, axis = 1)
merged.columns = list(X_train.columns) + list(y_train.columns)

merged = merged.dropna()
X_test = X_test.dropna()

#print(X_test.isnull().sum())

model.fit(merged.drop(columns = ['xyz_vaccine']),merged['xyz_vaccine'])
pred = model.predict_proba(X_test)
pred_xyz = pred[:,1]
```

```python
#for seasonal
X_test = test_feat.drop(columns = ['respondent_id'])
X_train = feat.drop(columns = ['respondent_id'])
y_train = targets.drop(columns = ['respondent_id','xyz_vaccine'])

merged = pd.concat([X_train,y_train],ignore_index = True, axis = 1)
merged.columns = list(X_train.columns) + list(y_train.columns)

merged = merged.dropna()
X_test = X_test.dropna()

#print(X_test.isnull().sum())

model.fit(merged.drop(columns = ['seasonal_vaccine']),merged['seasonal_vaccine'])
pred = model.predict_proba(X_test)
pred_seasonal = pred[:,1]
```
